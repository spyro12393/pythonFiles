```python
import numpy as np
import pandas as pd
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm, datasets
import matplotlib.pyplot as plt
```

```python
# Read Data
```

```python
zoo = pd.read_csv("zoo.csv")
zoo.head(5)
```

```python
# drop: 0:row, 1:column
zoo_modify = zoo.drop('name',1)
zoo_modify.head()
```

```python
zoo_modify_X, zoo_modify_y = zoo_modify.iloc[:,:16].values, zoo_modify.iloc[:, 16].values
train_X, test_X, train_y, test_y = train_test_split(zoo_modify_X, zoo_modify_y, test_size = 0.3)
```

```python
svm1 = SVC(kernel='linear', random_state=0, gamma=1.0, C=4.0)
svm1.fit(train_X, train_y)

test_y_predicted_linear = svm1.predict(test_X)

accuracy = metrics.accuracy_score(test_y, test_y_predicted_linear)
print(accuracy)
```

```python
svm2 = SVC(kernel='rbf', random_state=0, gamma=1.0, C=4.0)
svm2.fit(train_X, train_y)

test_y_predicted_rbf = svm2.predict(test_X)

accuracy = metrics.accuracy_score(test_y, test_y_predicted_rbf)
print(accuracy)
```

```python
svm3 = SVC(kernel='poly', random_state=0, gamma=1.0, C=4.0)
svm3.fit(train_X, train_y)

test_y_predicted_poly = svm3.predict(test_X)

accuracy = metrics.accuracy_score(test_y, test_y_predicted_poly)
print(accuracy)
```
